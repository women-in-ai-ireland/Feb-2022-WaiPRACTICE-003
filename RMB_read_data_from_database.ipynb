# Aim:
Provide easy access to pre-processed data to help speed up the analysis. Data is stored in a sqlite database that requires minimal effort to set up. User provides desired start, end date and station id and use.get_required_data() does all the rest

Note that the weather data is at an hourly level. The code takes the data and expands over the 12 5 min intervals in an hour with the assumption that each weather variable remains constant for that hour. The only exception, is the 'rain' variable - for this the code returns the average rain for each 5 min interval for the given hour i.e. total rain for the hour / 12

To use:
1. Download the zipped wai.db from GitHub
2. Unzip and place somewhere on your local computer. Make note of the location
3. Add that location to your code that pulls the data in e.g. in the *db_location* variable below.
4. Run *df = use.get_required_data(conn, start_date, end_date, station_id)*

And that is it.
